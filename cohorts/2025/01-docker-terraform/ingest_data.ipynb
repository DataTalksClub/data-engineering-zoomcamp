```python
import pandas as pd
from sqlalchemy import create_engine
```

### Taxi trip data

```python
!wget https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2019-10.csv.gz
```

```python
df = pd.read_csv('green_tripdata_2019-10.csv.gz')
df.head()
```

```python
# We need to provide a name for the table; we will use 'green_taxi_data'
print(pd.io.sql.get_schema(df, name='green_taxi_data'))
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
print(pd.io.sql.get_schema(df, name='green_taxi_data'))
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
engine.connect()
```

```python
# we can now use our engine to get the specific output for Postgres
print(pd.io.sql.get_schema(df, name='green_taxi_data', con=engine))
```

```python
df_iter = pd.read_csv('green_tripdata_2019-10.csv.gz', iterator=True, chunksize=100000)
df_iter
```

```python
df = next(df_iter)
df
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
df
```

```python
# we need to provide the table name, the connection and what to do if the table already exists
# we choose to replace everything in case you had already created something by accident before.
df.head(n=0).to_sql(name='green_taxi_data', con=engine, if_exists='replace')
```

```python
from time import time

while True: 
    try:
        t_start = time()
        df = next(df_iter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        
        df.to_sql(name='green_taxi_data', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
    except StopIteration:
        print('completed')
        break
```
